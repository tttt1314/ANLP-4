```python
#Rag model using sentence encoder and semantic search

import torch
from sentence_transformers import SentenceTransformer
import logging
import os
import numpy as np
from tqdm import tqdm

class FilteredAudioRAG:
    def __init__(self, audio_dir, csv_path, model_name='sentence-transformers/all-mpnet-base-v2'):
        logging.basicConfig(level=logging.INFO)
        self.logger = logging.getLogger(__name__)
        self.device = "cuda"

        # Load and filter data
        try:
            # Load the data
            self.data = pd.read_csv(csv_path)
            print(f"Initial dataset size: {len(self.data)}")

            # Print column names for verification
            print("\nAvailable columns:", self.data.columns.tolist())

            # Filter for is_audioset_eval = False
            self.data = self.data[self.data['is_audioset_eval'] == False]
            print(f"After audioset filtering: {len(self.data)}")

            # Add audio paths and filter for existing files
            self.data['audio_path'] = self.data['ytid'].apply(
                lambda x: os.path.join(audio_dir, f"{x}.wav")
            )
            self.data = self.data[self.data['audio_path'].apply(os.path.exists)]

            print(f"Final dataset size with existing audio files: {len(self.data)}")

            # # Sample a few captions for verification
            # print("\nSample captions from dataset:")
            # for idx, caption in enumerate(self.data['caption'].head(3)):
            #     print(f"{idx + 1}. {caption[:100]}...")

        except Exception as e:
            self.logger.error(f"Error loading data: {str(e)}")
            raise

        # Load model to GPU
        try:
            self.model = SentenceTransformer(model_name).to(self.device)
            self.logger.info(f"Loaded model on {self.device}")
        except Exception as e:
            self.logger.error(f"Error loading model: {str(e)}")
            raise

        self.caption_embeddings = None

    @torch.cuda.amp.autocast()
    def compute_embeddings(self):
        self.logger.info("Computing embeddings on GPU...")

        try:
            captions = self.data['caption'].tolist()

            # Compute embeddings
            embeddings = self.model.encode(
                captions,
                batch_size=32,
                show_progress_bar=True,
                convert_to_numpy=True,
                normalize_embeddings=True,
                device=self.device
            )

            # Convert to torch tensor and move to GPU
            self.caption_embeddings = torch.from_numpy(embeddings).to(self.device)

            self.logger.info(f"Computed embeddings tensor with shape: {self.caption_embeddings.shape}")

        except Exception as e:
            self.logger.error(f"Error computing embeddings: {str(e)}")
            raise

    @torch.cuda.amp.autocast()
    def retrieve_relevant_audio(self, query, top_k=2, similarity_threshold=0.5):
        self.logger.info(f"Processing query: {query}")

        if self.caption_embeddings is None:
            self.compute_embeddings()

        try:
            # Encode query and convert to tensor
            query_embedding = self.model.encode(
                query,
                convert_to_numpy=True,
                normalize_embeddings=True,
                device=self.device
            )
            query_embedding = torch.from_numpy(query_embedding).to(self.device)

            # Compute similarities on GPU
            similarities = torch.matmul(self.caption_embeddings, query_embedding)

            # Get top-k indices
            top_indices = torch.argsort(similarities, descending=True)[:top_k]

            results = []
            for idx in top_indices:
                idx = idx.item()
                if similarities[idx].item() >= similarity_threshold:
                    result = {
                        'youtube_id': self.data.iloc[idx]['ytid'],
                        'audio_path': self.data.iloc[idx]['audio_path'],
                        'caption': self.data.iloc[idx]['caption'],
                        'similarity_score': float(similarities[idx].cpu().numpy()),
                        'audioset_labels': self.data.iloc[idx]['audioset_positive_labels']
                    }
                    results.append(result)

            self.logger.info(f"Found {len(results)} results above threshold")
            return results

        except Exception as e:
            self.logger.error(f"Error in retrieval: {str(e)}")
            raise
```

```python
#test rag


def test_filtered_rag():
    """Test function to verify RAG system functionality"""
    print("\n=== Testing FilteredAudioRAG System ===\n")

    # Initialize RAG
    rag = FilteredAudioRAG(
        audio_dir='music_files/music_data',
        csv_path='musiccaps-public.csv'
    )

    # Test queries
    test_queries = [
        "electronic dance music with heavy bass",
        "peaceful piano melody",
        "rock music with guitar",
        "classical orchestra performance",
        "jazz saxophone solo"
    ]

    # Test each query
    for query in test_queries:
        print(f"\nQuery: {query}")
        print("-" * 50)

        results = rag.retrieve_relevant_audio(query, top_k=2)

        if results:
            for i, result in enumerate(results, 1):
                print(f"\nMatch {i}:")
                print(f"Caption: {result['caption'][:150]}...")
                print(f"Audio Path: {result['audio_path']}")
                print(f"Similarity Score: {result['similarity_score']:.4f}")
                print(f"YouTube ID: {result['youtube_id']}")
        else:
            print("No matches found above similarity threshold")

    print("\n=== Test Complete ===")

test_filtered_rag()
```

```python
#code to generate sample audio

import os
import torch
import torchaudio
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
import gc

# Initialize RAG system
audio_dir = "music_files/music_data"
csv_path = "musiccaps-public.csv"
rag = FilteredAudioRAG(audio_dir, csv_path)

# Change Text prompt Here
text_prompt = "Ragtime"

# Retrieve relevant audio paths
relevant_results = rag.retrieve_relevant_audio(text_prompt, top_k=1)
relevant_audio_paths = [result['audio_path'] for result in relevant_results]
print(f"Relevant audio paths: {relevant_audio_paths}")

# Clear memory
gc.collect()
torch.cuda.empty_cache()

# Initialize MusicGen Melody model
model = MusicGen.get_pretrained('facebook/musicgen-melody')  # Use the melody model
model.set_generation_params(duration=10)  # Generate 10 seconds

# Generate new audio using text prompt and individual reference audios
output_folder = './generated_audio'
os.makedirs(output_folder, exist_ok=True)

for i, audio_path in enumerate(relevant_audio_paths, start=1):
    # Load individual reference audio
    reference_waveform, sample_rate = torchaudio.load(audio_path)
    reference_waveform = torchaudio.functional.resample(reference_waveform, orig_freq=sample_rate, new_freq=32000)

    # Generate audio with individual reference
    wav = model.generate_with_chroma(
        [text_prompt],
        reference_waveform[None, :],
        melody_sample_rate=32000  # Set the reference waveform sample rate
    )
    generated_output_path = os.path.join(output_folder, text_prompt)
    audio_write(generated_output_path, wav[0].cpu(), model.sample_rate, strategy="loudness")
    print(f"Generated audio saved to: {generated_output_path}.wav")
```

```python
#gradio app Work in Progress

import gradio as gr
import os
import torch
import torchaudio
from audiocraft.models import MusicGen
from audiocraft.data.audio import audio_write
import gc

# Initialize RAG system
audio_dir = "music_files/music_data"  # Replace with your audio directory
csv_path = "musiccaps-public.csv"    # Replace with your CSV metadata file
rag = FilteredAudioRAG(audio_dir, csv_path)

# Initialize MusicGen Melody model
model = MusicGen.get_pretrained('facebook/musicgen-melody')
model.set_generation_params(duration=10)  # Generate 10 seconds of music

# Function to generate music
def generate_music(text_prompt, use_rag):
    try:
        gc.collect()
        torch.cuda.empty_cache()

        # Generate music without RAG
        if not use_rag:
            wav = model.generate([text_prompt])
            output_path_no_rag = f"/content/generated_no_rag"
            audio_write(output_path_no_rag, wav[0].cpu(), model.sample_rate, strategy="loudness")
            return output_path_no_rag, None  # Only no-RAG output

        # Generate music with RAG
        relevant_results = rag.retrieve_relevant_audio(query=text_prompt, top_k=1, similarity_threshold=0.5)
        relevant_audio_paths = [result['audio_path'] for result in relevant_results]

        if not relevant_audio_paths:
            return "No relevant audio found for RAG.", None

        # Load the reference audio
        audio_path = relevant_audio_paths[0]
        reference_waveform, sample_rate = torchaudio.load(audio_path)
        reference_waveform = torchaudio.functional.resample(reference_waveform, orig_freq=sample_rate, new_freq=32000)

        # Generate music with RAG
        wav_rag = model.generate_with_chroma(
            [text_prompt], reference_waveform[None, :], melody_sample_rate=32000
        )
        output_path_with_rag = f"/content/generated_with_rag"
        audio_write(output_path_with_rag, wav_rag[0].cpu(), model.sample_rate, strategy="loudness")

        return output_path_no_rag, output_path_with_rag

    except Exception as e:
        return f"Error: {str(e)}", None

# Gradio App
def app(text_prompt, use_rag):
    return generate_music(text_prompt, use_rag)

# Create Gradio Interface
with gr.Blocks() as demo:
    gr.Markdown("### Music Generation App with and without RAG")
    gr.Markdown("Provide a text prompt to generate music. Choose whether to use RAG for conditioning.")

    with gr.Row():
        text_input = gr.Textbox(label="Enter Text Prompt", placeholder="e.g., Jazz Music")
        rag_option = gr.Checkbox(label="Use RAG for Conditioning?", value=False)

    with gr.Row():
        submit_button = gr.Button("Generate Music")

    with gr.Row():
        output_no_rag = gr.Audio(label="Generated Music without RAG", type="filepath")
        output_with_rag = gr.Audio(label="Generated Music with RAG", type="filepath")

    # Bind functionality
    submit_button.click(
        fn=app,
        inputs=[text_input, rag_option],
        outputs=[output_no_rag, output_with_rag]
    )

# Launch the app
demo.launch(debug=True)
```
